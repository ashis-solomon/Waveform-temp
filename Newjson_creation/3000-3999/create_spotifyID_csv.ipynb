```python
import pandas as pd
import numpy as np
import json
import csv
import re
```

```python
with open('slice1.txt') as text:
    lines = text.readlines()
print(len(lines))

str = ""

for i in range(len(lines)):
    str += lines[i]
print(len(str))
    
```

```python
pattern = re.findall(r'spotify:track:+[\w\.-]+',str)
print(len(pattern))
print('Pattern list:', *pattern, sep='\n- ')
```

```python
df = pd.DataFrame(pattern) 
df.to_csv('track.csv') 
print(len(df))
```
